```python
#!/usr/bin/env python3
"""Amazon Business & Tech | dradabau
    File Converter From/To: CSV, Excel, Yaml, XML, JSON"""
import tkinter as tk
from tkinter import filedialog
from tkinter import *
#%pylab inline
#import pprint
import time
import random
import sys
import json
import pandas as pd
import yaml
import os
import re

def main():
    """Controller for the app"""
    
    userInput = ""
    print("**********************************************************")
    print("*****  FILE PARSER 1.0  *****")
    print("File parsing made easy.")
    print("Convert files From/To: CSV, Excel, Yaml, XML, JSON.")
    print("**********************************************************")

    while userInput == "":
        print("Are you ready to convert a file?")
        print("Select (y)es or (e)xit the program?")
        userInput = input("> ")
        if (userInput.lower() == 'e') or (userInput.lower()) == 'exit':
            break
            sys.exit("***Run program again when you wish to convert files***")
        elif userInput.lower() == 'y' or userInput.lower() == 'yes':
            continue
        else: userInput = ""
      
    
    print('Please select the file to convert from the pop up window.')
    # user selects file to convert to df, returns df
    df = selectfile() 
    # user selects conversion type, returns new_type
    new_type = pick_new_type() 
    #concats new_file_path + new_file_name + new_type, 
    #returns new_file_path
    new_file_path = make_new_file_path(new_type) 
    #Calls designated dfto* based on new_type
    write_file(new_type, new_file_path, df) 

    print("**********************************************************")
    print("The file has been converted.")
    print(" Your new file can be found at: \n", new_file_path)
    print("Thank you for using FileParser 1.0!!!!")
    

def selectfile(): 
    """Allows user to select the file to convert. Returns df """
    root = tk.Tk()
    root.withdraw()
    file = filedialog.askopenfilenames()
    fileString = ''.join(file)
   
    print("**********************************************************")
    endings =('xlsx', 'xml', 'yaml', 'json', 'csv')
    if fileString.endswith(endings):
        time.sleep(1)
        print("loading.")
        time.sleep(2 + random.random())
        print("loading..")
        time.sleep(random.random())
        print("loading...\n")
        print(f"Your file {fileString} has uploaded.")
    else: 
        print("You must select a(n) json, xml, yaml,\
        or csv file. Please try again.")
        time.sleep(3)
        main()
      
# Determine file type, read in file
    if fileString.endswith('yaml'):
        type = 'yaml'
        df = readYAML(fileString)
    elif fileString.endswith('xlsx'):
        type = 'xlsx'
        df = readExcel(fileString)
    elif fileString.endswith('json'):
        type = 'json'
        df = readJSON(fileString)
    elif fileString.endswith('csv'):
        type = 'csv'
        df = readCSV(fileString)
    elif fileString.endswith('xml'):
        type = 'xml'
        df = readXML(fileString)
    else:
        print("The file type is: UNKNOWN")
    print(f"This is a(n) {type} file.\n")
    return df

#    if fileString.endswith('json'):
#        readJSON(fileString)
#    elif fileString.endswith('csv'):
#        readCSV(fileString)
#    elif fileString.endswith('xlsx'):
#        readExcel(fileString)
#    elif fileString.endswith('yaml'):
#        readYAML(fileString)
#    elif fileString.endswith('xml'):
#        readXML(fileString)
#    return df

def readCSV(fileString):
    """CSV to df"""
    df = pd.read_csv(fileString)
    return df
    
def readExcel(fileString):
    """Excel to df"""
    df = pd.read_excel(fileString)
    return df

def readJSON(fileString):
    """JSON file to df"""
    #with open(f"{fileString}", "r") as js:
    #    jsonData = json.load(js)
    df = pd.read_json(fileString)
    return df

def readXML(fileString):
    """XML to df"""
    df = pd.read_xml(fileString)
    return df
    
def readYAML(fileString):
    """YAML to df"""
    with open(f"{fileString}", "r") as yf:
        yamlDict = yaml.load(yf, Loader=yaml.FullLoader)
    df = pd.DataFrame.from_dict(yamlDict)
    return df
    
def pick_new_type():
    """User picks new format"""
    print("\nWhat format would you like your file converted to?")
    print("***** MENU *****")
    print("1: CSV")
    print("2: Excel")
    print("3: JSON")
    print("4: XML")
    print("5: YAML")
    
    u_input = ''

    while u_input =='':
        new_type = ''
        u_input = str(input("\n> "))
        if u_input == '1':
            new_type = 'csv'
        elif u_input == '2':
            new_type = 'xlsx'
        elif u_input =='3':
            new_type = 'json'
        elif u_input == '4':
            new_type = 'xml'
        elif u_input == '5':
            new_type = 'yaml'
        else:
            #new_type = 'why does this show?'
            print("\nThe file type is: UNKNOWN\n")
            time.sleep(2)
            pick_new_type()   
        return new_type

def make_new_file_path(new_type):
    """Creates the filepath to write to"""
    new_type=new_type 
    # user selects filename string, returns new_file_name
    new_file_name = pick_new_file_name() 
    # user selects dirctory for new file, returns new_file_path
    new_file_path = pick_new_file_path()
    new_location = new_file_path + "/" + new_file_name + "." + new_type
    return new_location

def pick_new_file_name():
    """User designates file name for newly converted file"""
    new_file_name = ''
    while new_file_name =='':
        print("\nWhat would you like to name this file?")
        print("Please use letters, numbers and underscore(_) only.")
        new_file_name = str(input("> "))

        if re.match('[a-zA-Z0-9_]',new_file_name):
            print(f"{new_file_name}, is a good file name.")
            print("**********************************************************")
            return new_file_name
            break
            
        else:
            print("********* Invalid entry *********\n\
            Please use letters, numbers and underscore(_) only")
            time.sleep(1)
            pick_new_file_name()  
    
def pick_new_file_path():
    """User designates save path for converted file"""
    root = tk.Tk()
    root.withdraw()

    new_file_path =''
    while new_file_path =='':
        print("\nPick a folder to save the new file in: ")
        print("(check pop up window)")
        new_file_path = filedialog.askdirectory()
        if new_file_path !='':
            print("You chose: \n", new_file_path)
        while True:
            print("\nIs this the correct location for your new file? (y,n)")
            userValidation1 = input("> ")
            if userValidation1.lower() == 'y':
                return new_file_path
                break
            elif userValidation1.lower() == 'n':
                pick_new_file_path()  
        return new_file_path
    
    
def write_file(new_type, new_file_path, df):
    """Writes file to format/location designated by user"""
    if new_type == 'csv':
        dftoCSV(new_file_path,df)
    elif new_type == 'xlsx':
        dftoExcel(new_file_path,df)
    elif new_type == 'json':
        dftoJSON(new_file_path,df)
    elif new_type == 'xml':
        dftoXML(new_file_path,df)
    elif new_type == 'yaml':
        dftoYAML(new_file_path,df)    

def dftoCSV(new_file_path,df):
    """df to CSV"""
    df.to_csv(new_file_path, index=False)

def dftoExcel(new_file_path,df):
    """df to Excel"""
    df.to_excel(new_file_path, index=False)

def dftoJSON(new_file_path,df):
    """df to JSON"""
    df.to_json(new_file_path, orient="records")

def dftoXML(new_file_path,df):
    """converts df to XML"""
    df.to_xml(new_file_path)

def dftoYAML(new_file_path,df):
    """converts df to YAML"""
    with open(new_file_path, 'w') as file:
        yaml.dump({'result': df.to_dict(orient = 'records')}
                    , file, default_flow_style = False)
    
    

if __name__ == "__main__":
    main()
```
